```python
#K-means unsupervised clustering algorithm along with PCA implementation

from itertools import cycle
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from numpy.random import RandomState
import pylab as pl

def pca(X,n_components):
    pca = PCA(n_components=n_components, whiten=True).fit(X)
    X_pca = pca.transform(X)
    return X_pca

def plot(X,n_components,n_clusters, pca_flag):
    if pca_flag==True:
        X_pca = pca(X,n_components)
    else:
        X_pca = X
    #rng = RandomState(42)
    kmeans = KMeans(n_clusters=n_clusters, random_state=RandomState(42)).fit(X_pca)
    t = []
    i = 0
    while i<n_clusters:
        t.append("c"+str(i))
        i = i+1
    
    plot_2D(X_pca, kmeans.labels_, t)
    
def plot_c(X,n_components,n_clusters, pca_flag, column_name):
    if pca_flag==True:
        X_pca = pca(X,n_components)
    else:
        X_pca = X
    kmeans = KMeans(n_clusters=n_clusters, random_state=RandomState(42)).fit(X_pca)
    t = []
    i = 0
    while i<n_clusters:
        t.append("c"+str(i))
        i = i+1
    X[column_name] = kmeans.labels_
    plot_2D(X_pca, kmeans.labels_, t)
    return X

def plot_2D(data, target, target_names):
    colors = cycle('rgbcmykw')
    target_ids = range(len(target_names))
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names):
        pl.scatter(data[target == i, 0], data[target == i, 1],
                    c=c, label=label)
    pl.legend()
    pl.show()

```

```python
# Permutation and combination of all PCA values and number of clusters possible
#Select the best cluster to use for further analysis
#all graphs
t = []
k = 0
for i in range(2,7):
    for j in range(2,7):
        plot(df[['statuses_count','followers_count','friends_count', 'favourites_count','listed_count','tweet_freq']], i, j, True)
        k = k+1
```

```python
##Implementation of the Random Forest Classifier

def rfc(model_string,train,test):
    from patsy import dmatrices
    from sklearn import metrics
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import average_precision_score
    from sklearn.metrics import precision_score
    from sklearn.metrics import recall_score
    from sklearn.metrics import f1_score
    #from sklearn.metrics import accuracy_score


    ytest, Xtest = dmatrices(model_string,test, return_type="dataframe")
    # flatten y into a 1-D array
    ytest = (np.ravel(ytest)).astype(str)
    ytrain, Xtrain = dmatrices(model_string,train, return_type="dataframe")
    ytrain = (np.ravel(ytrain)).astype(str)

    model_et = RandomForestClassifier()
    model_et.fit(Xtrain, ytrain)
    # predict subscription for test set
    predicted_et = model_et.predict(Xtest)
    
    #print predicted_et
    
    # generate evaluation metrics
    #return predicted_et
    return predicted_et,metrics.accuracy_score(ytest,predicted_et),metrics.precision_score(ytest,predicted_et),metrics.recall_score(ytest,predicted_et),metrics.f1_score(ytest,predicted_et)
    #return [model_string,"Random Forest",metrics.r2_score(ytest,predicted_et),metrics.mean_absolute_error(ytest,predicted_et),metrics.mean_squared_error(ytest,predicted_et),metrics.median_absolute_error(ytest,predicted_et),metrics.explained_variance_score(ytest,predicted_et)]
```

```python
#clustser_grp3 is now the predicted cluster
model_string = "cluster_grp2~statuses_count+followers_count+friends_count+favourites_count+listed_count+tweet_freq"

test['cluster_grp3'],accuracy, precision,recall, f1  = rfc(model_string,train,test)
```
