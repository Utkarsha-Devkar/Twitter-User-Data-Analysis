```python
import pandas as pd
df5 = pd.read_csv("train_test_data6.csv")
print(len(df5))
df5.head(5)
#print(len(df))
```

```python
frames = [df1, df2, df3, df4, df5]

result = pd.concat(frames)
print(len(result))
result.describe()
```

```python

#to find the number of duplicates
pd.concat(g for _, g in result.groupby("user_id") if len(g) > 1)
```

```python
#This takes the last. Not the maximum though:
#result.drop_duplicates(cols='user_id', take_last=True)
results = result.drop_duplicates(['user_id'], keep='last')
print(len(results))
results.describe()
```

```python

results.to_csv("GroupedDataSet.csv",sep = ",", index=False)
```
